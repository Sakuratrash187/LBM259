# 2.2

Hier ist der Mode, Median und Variance von den Spalten max_temp, mean_temp und min_temp

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

temp_df = pd.read_csv(os.path.join('DatensatzLB259.csv'))
```

```python
temp_df.info()
temp_df.head()
```

```python
df = pd.DataFrame(temp_df)
description = df.describe()
print(description)

max_temp_mode = temp_df['max_temp'].mode()
# .median() und .var() (Varianz)
print(max_temp_mode)
mean_temp_mode = temp_df['mean_temp'].mode()
print(mean_temp_mode)
min_temp_mode = temp_df['min_temp'].mode()
print(min_temp_mode)
median_max_temp = temp_df['max_temp'].median()
print("median von max_temp", median_max_temp)
median_mean_temp = temp_df['mean_temp'].median()
print("median von mean_temp", median_mean_temp)
median_min_temp = temp_df['min_temp'].median()
print("median von min_temp", median_min_temp)
variance_max_temp = temp_df['max_temp'].var()
print("variance von max_temp", variance_max_temp)
variance_mean_temp = temp_df['mean_temp'].var()
print("variance von mean_temp", variance_mean_temp)
variance_min_temp = temp_df['min_temp'].var()
print("variance von min_temp", variance_min_temp)
```
